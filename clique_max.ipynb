```python
import numpy as np
import pandas as pd
import networkx as nx
from matplotlib import pyplot as plt
import warnings
import pickle
TRAIN_PATH = "./input/train.csv"
TEST_PATH = "./input/test.csv"
train = pd.read_csv(TRAIN_PATH, encoding="ISO-8859-1")
test = pd.read_csv(TEST_PATH, encoding="ISO-8859-1")
```

```python
with open('./node_cliques_len.pkl' ,'rb') as f:
    node_cliques_len = pickle.load(f)
```

```python
with open('./node_all_cliques.pkl' ,'rb') as f:
    node_all_cliques = pickle.load(f)
```

```python
def find_cliques_len(row):
    try:
        return node_cliques_len[row]
    except:
        return 0
```

```python
def get_longest_tuple_from_node_all_cliques(row):
    failed = 0
    questions = row['question1'], row['question2']
    clique_set = set()
    try:
        clique_set = clique_set | node_all_cliques[questions[0]]
    except:
        pass
    try:
        clique_set = clique_set | node_all_cliques[questions[1]]
    except:
        pass
    longest_clique = 0
    cliques = tuple()
    for each in clique_set:
        if questions[0] in each and questions[1] in each and len(each) > longest_clique:
            longest_clique = len(each)
            cliques = each
    return longest_clique
```

```python
def get_cliques_count(row):
    small_one = 0
    if row['q1_cliques_len'] > row['q2_cliques_len']:
        small_one = row['q1_cliques_len']
    else:
        small_one = row['q2_cliques_len']
    if small_one > 4:
        return 1
    else:
        return 0
```

```python
train['q1_cliques_len'] = train['question1'].apply(find_cliques_len)
train['q2_cliques_len'] = train['question2'].apply(find_cliques_len)
test['q1_cliques_len'] = test['question1'].apply(find_cliques_len)
test['q2_cliques_len'] = test['question2'].apply(find_cliques_len)
```

```python
train['get_cliques_count'] = train.apply(get_cliques_count, axis=1, raw=True)
test['get_cliques_count'] = test.apply(get_cliques_count, axis=1, raw=True)
```
